Start by defining two pdfs f & g

```python
def f(x):
    return np.exp(-0.5*x**2)/np.sqrt(2*np.pi)
```

```python
def g(x):
    sigma = .3
    return (0.5*np.exp(-(x-2)**2/(2*sigma*sigma))/np.sqrt(2*np.pi)/sigma
                +0.5*np.exp(-(x+3)**2/(2*sigma*sigma))/np.sqrt(2*np.pi)/sigma)
```

```python
#A histogram like example
def fhist(x):
    return (np.floor(x)+6)/sum(range(11))
def ghist(x):
    return 0.2-(np.floor(x)+6)/sum(range(11))
#(f,g)=(fhist,ghist) #uncomment for histogram like example
```

```python
g(-4.9)
```

```python
num = 200
xarray = np.linspace(-5,5,num)
```

It's going to be easier to work in the space of the sqrt of distributions, so let's define a simple transformation q 

```python
q = lambda h: (lambda x: np.sqrt(h(x)))
```

```python
q(f)(0)
```

Now we define the inner product on the distributions

```python
def inner_product(q1,q2, xmin, xmax):
    # this is a simple numeric integration
    num = 200.
    xarray = np.linspace(xmin,xmax,num)
    return np.sum(q1(xarray) * q2(xarray))*(xmax-xmin)/num
```

Test inner product, should be close to (1,1) b/c pdfs are normalized

```python
(inner_product(q(g),q(g),-5,5), inner_product(q(f),q(f),-5,5))
```

```python
fdotg = inner_product(q(f),q(g),-5,5)
fdotg
```

```python
# hard-coded fdotg here, not general
tangent_dir = lambda q1, q2: (lambda x: q2(x) - fdotg*q1(x))
```

```python
tangent_ftog = tangent_dir(q(f),q(g))
```

```python
tangent_ftog_norm = np.sqrt(inner_product(tangent_ftog, tangent_ftog, -5,5))
tangent_ftog_norm
```

Plot the direction that is tangent to f in direction of g (in the sqrt(pdf) space)


```python
fig = plt.figure(figsize=(5*3,4.5))
plt.plot(xarray,tangent_dir(f,g)(xarray))
```

Now the cool part, here is the geodesic in the Q space:

```python
geodesic = lambda t: (lambda x: q(f)(x)*np.cos(t)
                      +tangent_ftog(x)*np.sin(t)/tangent_ftog_norm)
```

Square it to get a distribution

```python
geodesic_dist = lambda t: (lambda x: (geodesic(t)(x))**2)
```

The geodesic starts at f for t=0 and gets to g at  t=arccos( fdotg) 

```python
t_g = np.arccos(fdotg)
t_g
```

check geodesic_dist(t=0) = f, geodesic_dist(t_g) = g.  Should be ~(0,0)

```python
(geodesic_dist(0)(0)-f(0), geodesic_dist(t_g)(0)-g(0))
```

```python
fig = plt.figure(figsize=(5*3,4.5))
for t in np.linspace(0,t_g, 7):
    plt.plot(xarray,geodesic_dist(t)(xarray))            
```

## 2d interpolation

Same idea, but we will start at f and have two tangent directions based on g1 and g2. One can choose any linear combination of these directions, parametrized by theta, to define a new geodesic.

```python
def g1(x):
    sigma = .3
    return np.exp(-(x-2)**2/(2*sigma*sigma))/np.sqrt(2*np.pi)/sigma
def g2(x):
    sigma = .3
    return np.exp(-(x+2)**2/(2*sigma*sigma))/np.sqrt(2*np.pi)/sigma
           
```

```python
fdotg1 = inner_product(q(f),q(g1),-5,5)
fdotg2 = inner_product(q(f),q(g2),-5,5)
(fdotg1,fdotg2)
```

```python
tangent_tog1 = lambda x: q(g1)(x) - fdotg2*q(f)(x)
tangent_tog2 = lambda x: q(g2)(x) - fdotg2*q(f)(x)
```

```python
tangent_theta = lambda theta: lambda x: np.cos(theta)*tangent_tog1(x) + np.sin(theta)*tangent_tog2(x)
```

```python
tangent_theta_norm = lambda theta: np.sqrt(inner_product(tangent_theta(theta), tangent_theta(theta), -5,5))
(tangent_theta_norm(0),tangent_theta_norm(np.pi/2), tangent_theta_norm(np.pi/4))
```

```python
geodesic_2d = lambda t,theta: (lambda x: q(f)(x)*np.cos(t)
                      +tangent_theta(theta)(x)*np.sin(t)/tangent_theta_norm(theta))
```

```python
t_g1 = np.arccos(inner_product ( q(f), q(g1), -5, 5 ) )
t_g1
```

```python
geodesic_theta_dist = lambda t,theta: lambda x: (geodesic_2d(t,theta)(x))**2
```

```python
fig = plt.figure(figsize=(5*3,4.5))
plt.plot(xarray,g1(xarray))
for t in np.linspace(0,t_g1, 3):
    for theta in np.linspace(0,np.pi/2, 3):
        plt.plot(xarray,geodesic_theta_dist(t,theta)(xarray))            
```

```python
fig = plt.figure(figsize=(5*3,4.5))
for t in np.linspace(0,t_g1, 7):
    plt.plot(xarray,geodesic_theta_dist(t,np.pi/10)(xarray))            
```

## Alternative interpolation (harder to generalize)

```python
g1dotg2 = inner_product(q(g1),q(g2),-5,5)
```

```python
tangent_g1tog2 = lambda x: q(g2)(x) - g1dotg2*q(g1)(x)
```

```python
tangent_g1tog2_norm = np.sqrt(inner_product(tangent_g1tog2, tangent_g1tog2, -5,5))
tangent_g1tog2_norm
```

```python
t_g1_g2 = np.arccos(g1dotg2)
```

```python
#interpolation parameter theta so that theta=pi/2 is g2
g_theta = lambda theta: (lambda x: q(g1)(x)*np.cos(theta*t_g1_g2*2/np.pi)
                      +tangent_g1tog2(x)*np.sin(theta*t_g1_g2*2/np.pi)/tangent_g1tog2_norm)
```

```python
g_theta_dist = lambda theta: lambda x: (g_theta(theta)(x))**2
```

```python
for theta in np.linspace(0,np.pi/2,5):
    plt.plot(xarray,g_theta_dist(theta)(xarray))
```

```python
fdotgtheta = lambda theta: inner_product(q(f), g_theta(theta), -5,5)
```

```python
tangent_theta_b = lambda theta: lambda x: g_theta(theta)(x) - fdotgtheta(theta)*q(f)(x)
```

```python
tangent_theta_b_norm = lambda theta: np.sqrt(inner_product(tangent_theta_b(theta), tangent_theta_b(theta), -5,5))
tangent_theta_b_norm(0)
```

```python
geodesic_2d_b = lambda t,theta: (lambda x: q(f)(x)*np.cos(t)
                      +tangent_theta_b(theta)(x)*np.sin(t)/tangent_theta_b_norm(theta))
```

```python
geodesic_theta_b_dist = lambda t,theta: lambda x: (geodesic_2d_b(t,theta)(x))**2
```

```python
fig = plt.figure(figsize=(5*3,4.5))
plt.plot(xarray,g1(xarray))
for t in np.linspace(0,t_g1, 3):
    for theta in np.linspace(0,np.pi/2, 3):
        plt.plot(xarray,geodesic_theta_b_dist(t,theta)(xarray))   
        plt.plot(xarray,geodesic_theta_dist(t,theta)(xarray))            
```

```python
fig = plt.figure(figsize=(5*3,4.5))
for t in np.linspace(0,t_g1, 3):
    for theta in np.linspace(0,np.pi/2, 3):
        plt.plot(xarray,geodesic_theta_b_dist(t,theta)(xarray)-geodesic_theta_dist(t,theta)(xarray))            
```
