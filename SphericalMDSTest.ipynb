```python
import numpy as np
import numpy.linalg as linalg
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn import manifold
from mpl_toolkits.mplot3d import Axes3D
```

```python
dim = 3 # or any positive integer
nPoints = dim*10
pointsOnSphere = np.random.normal(size=(dim, nPoints)) 
#x /= numpy.linalg.norm(x, axis=1)[:, numpy.newaxis] #need newer scipy
norm = np.sqrt(np.sum(pointsOnSphere**2, 0))
pointsOnSphere /= norm
pointsOnSphere = pointsOnSphere.T
```

```python
def getDistance(x,y):
    return np.sqrt(np.sum((x-y)**2, 0))
```

```python
#Test
getDistance( np.array([2,2,2]), np.array([1,1,1]))-np.sqrt(3)
```

```python
tempSim=[]
for x in pointsOnSphere:
    temp = []
    for y in pointsOnSphere:
        temp.append(getDistance(x,y))
    tempSim.append(temp)
distances=np.array(tempSim)
```

```python
seed = np.random.RandomState(seed=3)
mds = manifold.MDS(n_components=dim, metric=True, max_iter=3000, eps=1e-9, random_state=seed,dissimilarity="precomputed", n_jobs=1)
embed3d = mds.fit(distances).embedding_
```

```python
fig = plt.figure(figsize=(5,5))
subpl = fig.add_subplot(111,projection='3d')
subpl.scatter(embed3d[:, 0], embed3d[:, 1], embed3d[:, 2])
subpl.view_init(42, 101) #looks good when njobs=-1
subpl.view_init(-130,-33)#looks good when njobs=1
plt.show()
```

## Try to align point to x_n = -1 

From stack exchange: If u and v are two orthonormal vectors, a matrix that rotates the span of u and v by angle θ is A=I+sin(θ)(vuT−uvT)+(cos(θ)−1)(uuT+vvT). 

   * u can be (0,...,-1) and can build the orthonormal v in the direction of x by doing (x - u<u,x>)/norm.

```python
u = np.zeros(dim)
u[-1]=-1
v = pointsOnSphere[0] - pointsOnSphere[0].dot(u)*u
v/=linalg.norm(v)
v.dot(u)
```

```python
theta = -np.arccos(pointsOnSphere[0].dot(u))
```

```python
linalg.norm(u)
```

```python
(np.outer(v,u.T), np.outer(v,u), np.outer(v.T,u))
vuT=np.outer(v,u.T)
```

```python
uvT = np.outer(u,v)
```

```python
uuT = np.outer(u,u)
```

```python
vvT = np.outer(v,v)
```

```python
I = np.identity(dim)
```

```python
A=I + np.sin(theta)*(vuT-uvT) + (np.cos(theta)-1.)*(uuT+vvT)
```

```python
A
```

```python
A.dot(pointsOnSphere[0])
```

```python
#there must be a better way
rotatedPointsOnSphere = pointsOnSphere.copy()
for i, point in enumerate(pointsOnSphere):
    #if i==0: continue
    rotatedPointsOnSphere[i] = A.dot(point)
    if rotatedPointsOnSphere[i,2]>0:
        rotatedPointsOnSphere[i,2] *= -1
rotatedPointsOnSphere[0] = np.zeros(dim)
rotatedPointsOnSphere[0,-1] = -1.
rotatedPointsOnSphere[:3]
```

```python
gnomonicProjection = pointsOnSphere.copy()
for i, point in enumerate(rotatedPointsOnSphere):
    gnomonicProjection[i] = -1.*point/point[-1]
```

```python
#some more points on the sphere to help with visualization
markers = np.random.normal(size=(dim, nPoints*10)) 
norm = np.sqrt(np.sum(markers**2, 0))
markers /= norm
markers = markers.T
```

```python
nSamples = 15 #samples along ray for visualization
rays = np.zeros((nPoints*nSamples,dim))
for i, point in enumerate(gnomonicProjection):
    for j, c in enumerate(np.linspace(0,1,nSamples)):
        rays[i*nSamples+j] = c * point
```

```python
from scipy.spatial import Delaunay
```

```python
gnomonicSimplex = Delaunay(gnomonicProjection[:dim,:dim-1])
```

```python
example = 0
exIndex = 0 #careful about indexing
for i, point in enumerate(gnomonicProjection[:,:dim-1]):
    if i > dim and gnomonicSimplex.find_simplex(point)==0:
        example = point
        exIndex = i
        break
example, exIndex, gnomonicProjection[exIndex,:dim-1]
```

```python
baryCoords = gnomonicSimplex.transform[0,:dim-1,:].dot(example)
```

```python
normedVertices = gnomonicProjection[:3,:2]
normedVertices[1] /= linalg.norm(normedVertices[1])
normedVertices[2] /= linalg.norm(normedVertices[2])
normedSimplex = Delaunay(normedVertices)
normedBaryCoords = normedSimplex.transform[0,:dim-1,:].dot(example)
normedBaryCoords
```

```python
correctTangent = rotatedPointsOnSphere[exIndex] - rotatedPointsOnSphere[0] * (rotatedPointsOnSphere[exIndex].dot(rotatedPointsOnSphere[0]))
correctTangent = rotatedPointsOnSphere[exIndex] - rotatedPointsOnSphere[0] * (rotatedPointsOnSphere[0].dot(rotatedPointsOnSphere[exIndex]))
correctTangent /= linalg.norm(correctTangent)
correctTangent, rotatedPointsOnSphere[exIndex]/linalg.norm(rotatedPointsOnSphere[exIndex,:dim-1])
```

```python
t1 = rotatedPointsOnSphere[1] - rotatedPointsOnSphere[0] * (rotatedPointsOnSphere[1].dot(rotatedPointsOnSphere[0]))
t1 /= linalg.norm(t1)
#t1 *= np.arccos(rotatedPointsOnSphere[1].dot(rotatedPointsOnSphere[0]))
t2 = rotatedPointsOnSphere[2] - rotatedPointsOnSphere[0] * (rotatedPointsOnSphere[2].dot(rotatedPointsOnSphere[0]))
t2 /= linalg.norm(t2)
#t2 *= np.arccos(rotatedPointsOnSphere[2].dot(rotatedPointsOnSphere[0]))
np.matrix([t1,t2])
invT = np.zeros(2*dim).reshape(2,dim)
invT[0] = t1
invT[1] = t2
invT
#tangent = baryCoords[0]*t1 + baryCoords[1]*t2 #wrong
#tangent = invT.T.dot(normedBaryCoords[::-1])
#tangent/=linalg.norm(tangent)
tangent = invT.T.dot(normedBaryCoords) #not sure about order of baryCoords
tangent/=linalg.norm(tangent)
tangent, correctTangent
```

Yeah it works!!!!
